## Replicating a Jupyter Notebook through a JSON file 

```python
import json
from pprint import pprint
import pandas as pd
from IPython.display import display, Markdown, Code
```

 ⚠️  *YOU NEED TO RETREIVE THE JSON FILE FROM DEVELOPERS TOOLS BEFOREHAND*

```python
# replace 'path_to_json_file' with the actual path to your JSON file
with open('/Users/elenagarciamanes/Documents/Resources/l2-guidelines.json', 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
df = pd.json_normalize(data['content']['cells'])
display(df)
```

### *REPLICA* OF THE NOTEBOOK

Replicating the content of the Jupyter Notebook stored in the JSON file by making use the the 'cell_type' column:

```python
for index, row in df.iterrows():
    if row["cell_type"] == "markdown":
        display(Markdown(row["source"]))
    elif row["cell_type"] == "code":
        display(Code(row["source"], language="python"))
```
